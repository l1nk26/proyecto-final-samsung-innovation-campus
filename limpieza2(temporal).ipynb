```python
import pandas as pd
import numpy as np
import re
import ast
pd.set_option('display.max_colwidth', 300)
pd.set_option('display.max_columns', None)
```

```python
## funciones

# Para links

# elimina cualquier cosa que este antes del https
def corregir_link(enlace):
    return re.sub(r'.*?(https://)', r'\1', enlace, flags=re.DOTALL)

# arregla bien el enlace
def corregir_links(links):
    if links.startswith("c("):
        links = links[2:-1]
    links = links.replace('"', '')
    links = links.replace("'", '')
    links = links.replace("[", '')
    links = links.replace("]", '')
    links = links.split(', ')
    # a veces vienen enlaces con regalitos antes del https por lo que hay que eliminar eso
    links = [corregir_link(link) for link in links]
    return links

# Para instrucciones

def eliminar_antes_mayuscula(cadena):
    # expresión regular para encontrar y reemplazar cualquier cosa antes de una letra mayúscula
    return re.sub(r'.*?([A-Z])', r'\1', cadena, count = 1, flags=re.DOTALL)

def corregir_instruccion(instr):
    if instr.startswith("c("):
        instr = instr[2:-1]
    instr = instr.replace('"', '')
    instr = instr.replace('\n', '')
    instr = instr.replace("'", '')
    instr = instr.replace("]", '')
    instr = instr.replace("[", '')
    instr = instr.split(', ')
    instr = [eliminar_antes_mayuscula(cInstr) for cInstr in instr] # la cInstr significa cada instruccion
    return instr


# Para Tags

def repararx2(palabras):
    palabras = palabras.replace("']", '')
    palabras = palabras.replace("['", '')
    palabras = palabras.split("', '")
    return palabras

no_admitido = ['ajun', 'alloween', 'amping', 'anadian', 'anning',
               'anukkah', 'aribbean','asy','avory','awaiian','ealthy'
               'eans','eanut Butter','eat', 'eef Barley Soup','eeknight'
               'eep Fried','efrigerator','egan','egetable','ery Low Carbs'
               'essert','everages','ew Zealand','exican','eynolds Wrap Contest'
               'frican','hanksgiving','heese','hristmas','ice','id Friendly'
               'igh In...','inter','irthday','ish Halibut','mall Appliance'
               'ndian','ne Dish Meal','nexpensive','o Cook','ollege Food'
               'ookie & Brownie','or Large Groups','ortilla Soup','osher'
               'otluck','ousehold Cleaner','outh American','outhwestern U.S.'
               'ow Cholesterol','ow Protein','paghetti','picy','pring',
               'rab','rains','reads','reakfast','ree Of...','reezer','reole' 
               'rom Scratch','rout','ruit','runch','sian','t. Patrick\'s Day'
               'team','tew','tir Fry','tove Top','ummer','una','unch/Snacks'
               'uropean','uts','ven','weet','ysters'] 


def actualizar_lista(lista, elemento):
    if elemento not in lista:
        lista.append(elemento)
    if no_admitido in lista:
        lista.remove(no_admitido)
    return lista


# Para Tags

def convertir_a_lista(cadena):
    try:
        return ast.literal_eval(cadena)
    except (ValueError, SyntaxError):
        return []  # Devolver una lista vacía si hay un error en la cadena


# Para Tags
def actualizar_mins(elemento):
    # Definir los valores a reemplazar y sus sustitutos
    cambios = {
        ' 15 Mins': '15 Mins or less',
        ' 30 Mins': '30 Mins or less',
        ' 4 Hours': '4 Hours or less',
        ' 60 Mins': '60 Mins or less',
        '< 15 Mins': '15 Mins or less',
        '< 30 Mins': '30 Mins or less',
        '< 4 Hours': '4 Hours or less',
        '< 60 Mins': '60 Mins or less'
    }
    
    # Si el elemento es una lista, aplicamos cambios a cada item
    if isinstance(elemento, list):
        return [cambios.get(item, item) for item in elemento]
    # Si es un string, aplicamos el cambio directamente
    elif isinstance(elemento, str):
        return cambios.get(elemento, elemento)
    # Para otros tipos (ej. NaN), retornamos el mismo valor
    else:
        return elemento
    

# funcion base para una funcion que no he hecho
# def delete_mins(elemento):
#     # Definir los valores a reemplazar y sus sustitutos
#     cambios = {
#         ' 15 Mins': '15 Mins or less',
#         ' 30 Mins': '30 Mins or less',
#         ' 4 Hours': '4 Hours or less',
#         ' 60 Mins': '60 Mins or less',
#         '< 15 Mins': '15 Mins or less',
#         '< 30 Mins': '30 Mins or less',
#         '< 4 Hours': '4 Hours or less',
#         '< 60 Mins': '60 Mins or less'
#     }


# Para el final
def corregir(links, i = False):
    if links.startswith("c("):
        links = links[2:-1]
    links = links.replace('"', '')
    links = links.replace("'", '')
    links = links.replace("[", '')
    links = links.replace("]", '')
    if i:
        links = links.replace(', ',',\n')
        links = links.split(',')
    else:
        links = links.split(', ')
    # a veces vienen enlaces con regalitos antes del https por lo que hay que eliminar eso
    #links = [corregir_link(link) for link in links]
    return links
```
