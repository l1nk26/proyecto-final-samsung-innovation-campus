```python
# elimina cualquier cosa que este antes del https
def corregir_link(enlace):
    return re.sub(r'.*?(https://)', r'\1', enlace, flags=re.DOTALL)

# arregla bien el enlace
def corregir_links(links):
    if links.startswith("c("):
        links = links[2:-1]
    links = links.replace('"', '')
    links = links.replace("'", '')
    links = links.replace("[", '')
    links = links.replace("]", '')
    links = links.split(', ')
    # a veces vienen enlaces con regalitos antes del https por lo que hay que eliminar eso
    links = [corregir_link(link) for link in links]
    return links
    
```

```python
data_procesada['Images'] = data_procesada['Images'].apply(lambda links: corregir_links(links))
data_procesada['Images'].iloc[0]
```

```python
# se llaman link porque me da flojera usar ctrl f y cambiar su nombre
def eliminar_antes_mayuscula(cadena):
    # expresión regular para encontrar y reemplazar cualquier cosa antes de una letra mayúscula
    return re.sub(r'.*?([A-Z])', r'\1', cadena, count = 1, flags=re.DOTALL)

def corregir_instruccion(links):
    if links.startswith("c("):
        links = links[2:-1]
    links = links.replace('"', '')
    links = links.replace('\n', '')
    links = links.replace("'", '')
    links = links.replace("]", '')
    links = links.replace("[", '')
    links = links.split(', ')
    links = [eliminar_antes_mayuscula(link) for link in links]
    return links
```

#### Unificar categorias y palabras clave

```python
def repararx2(palabras):
    palabras = palabras.replace("']", '')
    palabras = palabras.replace("['", '')
    palabras = palabras.split("', '")
    return palabras

no_admitido = ['ajun', 'alloween', 'amping', 'anadian', 'anning',
               'anukkah', 'aribbean','asy','avory','awaiian','ealthy'
               'eans','eanut Butter','eat', 'eef Barley Soup','eeknight'
               'eep Fried','efrigerator','egan','egetable','ery Low Carbs'
               'essert','everages','ew Zealand','exican','eynolds Wrap Contest'
               'frican','hanksgiving','heese','hristmas','ice','id Friendly'
               'igh In...','inter','irthday','ish Halibut','mall Appliance'
               'ndian','ne Dish Meal','nexpensive','o Cook','ollege Food'
               'ookie & Brownie','or Large Groups','ortilla Soup','osher'
               'otluck','ousehold Cleaner','outh American','outhwestern U.S.'
               'ow Cholesterol','ow Protein','paghetti','picy','pring',
               'rab','rains','reads','reakfast','ree Of...','reezer','reole' 
               'rom Scratch','rout','ruit','runch','sian','t. Patrick\'s Day'
               'team','tew','tir Fry','tove Top','ummer','una','unch/Snacks'
               'uropean','uts','ven','weet','ysters'] 


def actualizar_lista(lista, elemento):
    if elemento not in lista:
        lista.append(elemento)
    if no_admitido in lista:
        lista.remove(no_admitido)
    return lista
```

```python
columns = ['RecipeCategory','Keywords']

data_procesada['RecipeCategory'] = data_procesada.apply(lambda row: actualizar_lista(repararx2(row['Keywords']), row['RecipeCategory']), axis = 1)

#repararx2(data_pruebas[columns[1]].iloc[1]) + repararx2(data_pruebas[columns[0]].iloc[1])

```
