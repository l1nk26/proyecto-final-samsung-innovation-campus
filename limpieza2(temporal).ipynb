```python
# elimina cualquier cosa que este antes del https
def corregir_link(enlace):
    return re.sub(r'.*?(https://)', r'\1', enlace, flags=re.DOTALL)

# arregla bien el enlace
def corregir_links(links):
    if links.startswith("c("):
        links = links[2:-1]
    links = links.replace('"', '')
    links = links.replace("'", '')
    links = links.replace("[", '')
    links = links.replace("]", '')
    links = links.split(', ')
    # a veces vienen enlaces con regalitos antes del https por lo que hay que eliminar eso
    links = [corregir_link(link) for link in links]
    return links
    
```

```python
data_procesada['Images'] = data_procesada['Images'].apply(lambda links: corregir_links(links))
data_procesada['Images'].iloc[0]
```

```python
# se llaman link porque me da flojera usar ctrl f y cambiar su nombre
def eliminar_antes_mayuscula(cadena):
    # expresión regular para encontrar y reemplazar cualquier cosa antes de una letra mayúscula
    return re.sub(r'.*?([A-Z])', r'\1', cadena, count = 1, flags=re.DOTALL)

def corregir_instruccion(links):
    if links.startswith("c("):
        links = links[2:-1]
    links = links.replace('"', '')
    links = links.replace('\n', '')
    links = links.replace("'", '')
    links = links.replace("]", '')
    links = links.replace("[", '')
    links = links.split(', ')
    links = [eliminar_antes_mayuscula(link) for link in links]
    return links
```

```python
#Vamo a desmenusarlo :inserte cara moai
def convertir_a_lista(cadena):
    try:
        return ast.literal_eval(cadena)
    except (ValueError, SyntaxError):
        return []  # Devolver una lista vacía si hay un error en la cadena
C['tags'] = C['tags'].apply(convertir_a_lista)
allTags = [item for sublist in C['tags'] for item in sublist ]
tags = list(set(allTags))
```

```python
tags.sort()
for tag in tags:
    print(tag,"\n")
```

```python
# un cambio de variable nunca viene mal
lista = tags
```

```python
# idx es el indice en donde los valores se ponen raros
idx = next((i for i, x in enumerate(lista) if x > 'Z'), len(lista))

tags_final = lista[:idx]

for tag in tags_final:
    print(tag,"\n")
tags_final = pd.DataFrame({'tags_unique': tags_final})
    
```
